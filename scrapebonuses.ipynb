```python
#!/usr/bin/env python3
# coding: utf-8


import requests

headers = requests.utils.default_headers()
headers.update({ 'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:52.0) Gecko/20100101 Firefox/52.0'})


from bs4 import BeautifulSoup

for year in range(2000,2020):
    url = "https://www.moneycontrol.com/stocks/marketinfo/bonus/homebody.php?sel_year="+str(year)
    req = requests.get(url,headers)
    soup = BeautifulSoup(req.content, 'html.parser')

    values = []
    dict = {}

    for x in soup.find_all(class_="dvd_brdb"):
        values.append(x.get_text())

    del values[0:6] #headers
    print("values length :", len(values))
    i=0

    dict = {}

    while(i<len(values)):
        dict[values[i]] = [values[i+1], values[i+2], values[i+3], values[i+4]]
        i+=5

    print('init dict :', dict)

    links = []

    for x in soup.find_all("a", class_="bl_12"):
        links.append(x.get('href'))

    del links[0]
    del links[-1]
    links = list(dict.fromkeys(links))
    print(links)


    print(len(links), "dict : ", len(dict.keys()))

    NSE = []
    i=0
    while(i<len(links)):
        url = "http://www.moneycontrol.com"+links[i]
        req = requests.get(url,headers)
        if(req.status_code >=400):
            i+=1
            print('some error at :', i)
        else:        
            newsoup = BeautifulSoup(req.content, 'html.parser')    

            p = newsoup.find(class_="bsns_pcst")
            span = p.find_all("span")

            counter=0
            for y in span:
                if(counter==1):
                    NSE.append(y.get_text())
                counter+=1
            i+=1

        print(i)
        #print(NSE)

    print("NSE Names : ", len(NSE), NSE)

    print("----------------FINAL----------------\n")

    print(len(NSE), NSE)
    print(len(dict), dict.keys())


    i=0
    with open('D:\\Stock-Market-Analysis\\bonusData\\'+str(year)+'bonuses.txt', 'w+') as f:
        for key, value in dict.items():          
            f.write('%s,%s,%s,%s,%s,%s\n' % (key, NSE[i], value[0],value[1],value[2],value[3]))
            i+=1

    f.close()
```
